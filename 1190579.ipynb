```python
import pandas as pd
df = pd.concat([df_island_Torgersen, df_island_Biscoe, df_island_Dream], ignore_index=True)
df.info()
```

```python
# Import the necessary library for one-hot encoding
from sklearn.preprocessing import OneHotEncoder

# Create an instance of the OneHotEncoder with 'handle_unknown' set to 'ignore'
enc = OneHotEncoder(handle_unknown='ignore')

# Columns to one-hot encode
columns_to_encode = ['sex', 'species', 'island']

# Create an empty DataFrame to store the one-hot encoded columns
df_encoded = pd.DataFrame()

# Fit and transform each specified column into one-hot encoded arrays
for col in columns_to_encode:
    encoded_array = enc.fit_transform(df[[col]]).toarray()
    # Create a DataFrame from the one-hot encoded array
    df_encoded = pd.concat([df_encoded, pd.DataFrame(encoded_array, columns=enc.get_feature_names_out([col]))], axis=1)

# Create a new DataFrame with the original columns and the one-hot encoded columns
df_combined = pd.concat([df, df_encoded], axis=1)

# Drop the original columns that were one-hot encoded
df_combined.drop(columns=columns_to_encode, inplace=True)

# Display the first few rows of the new DataFrame
print(df_combined.head())
```

```python
from sklearn.feature_selection import VarianceThreshold

# Assuming df_combined is your DataFrame with one-hot encoded columns

# Initialize the VarianceThreshold selector with a threshold
selector = VarianceThreshold(threshold=0.3)

df_combined_copy = df_combined.copy()

# Fit the selector to the data
selector.fit(df_combined)

# Get the mask of selected features
selected_features_mask = selector.get_support()

# Apply the selector to the data and create a new DataFrame
df_high_variance = df_combined.loc[:, selected_features_mask]

df_high_variance["species_Adelie"] = df_combined_copy["species_Adelie"]
df_high_variance["species_Chinstrap"] = df_combined_copy["species_Chinstrap"]
df_high_variance["species_Gentoo"] = df_combined_copy["species_Gentoo"]

# Variance of the data
selected_variances = selector.variances_

# Display the variances
print("Variances of each feature after applying the threshold:")
print(selected_variances)

print("\nData after applying threshold=0.3:")
print(df_high_variance.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to the selected features
scaler.fit(df_high_variance)

# Transform the selected features using the scaler
data_scaled = scaler.transform(df_high_variance)

# Create a new DataFrame with the scaled data
df_high_variance_scaled = pd.DataFrame(data_scaled, columns=df_high_variance.columns)

# Display the summary statistics of the scaled DataFrame
print(df_high_variance_scaled.describe())
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df_high_variance_scaled.drop(columns=['species_Adelie', 'species_Gentoo', 'species_Chinstrap'])
y = df_high_variance_scaled[['species_Adelie', 'species_Gentoo', 'species_Chinstrap']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Create PCA instance and fit to the data
pca = PCA(n_components=3) # Specify the number of components
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
# Train a classifier on the retained PCA components and evaluate
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_pca, y_train)
y_pred = clf.predict(X_test_pca)
accuracy_variance = accuracy_score(y_test, y_pred)
# Print the explained variance ratio for each selected component
print(f"Explained variance ratio for each PCA component: {pca.explained_variance_ratio_}")
print(f"Number of original features: {X_train.shape[1]}")
print(f"Number of features retained after PCA: {X_train_pca.shape[1]}")
print(f"Accuracy after PCA (testing accuracy): {accuracy_variance}")
```
